<h1> #1 Analyze the dataset and clean it. Preapare it for task

```python
data.info()
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
prices = list(data['Price'])
non_numeric = []
for i,val in enumerate(prices):
    try:
        int(val)
    except:
        non_numeric.append(val)
        print(val, end=" ")
```

```python
# Dropping POA from dataset
for non_num in non_numeric:
    data = data[data['Price'] != non_num]
    
# Converting price column from object to int type
data['Price_int'] = data['Price'].apply(lambda x: int(x))
data.info()
```
